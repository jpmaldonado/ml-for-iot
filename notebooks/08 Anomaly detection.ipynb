```python
# Import the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# To ensure the results are reproducible
np.random.seed(7)

salaries = []
for _ in range(100):
    salary = np.random.randint(1000,2500)
    salaries.append(salary)

```

```python
salaries[16] = 23
salaries[65] = 17
salaries = np.array(salaries)
salaries = salaries.astype('float64')
```

```python
from scipy.cluster.vq import kmeans

centroids, avg_distance = kmeans(salaries, 4)
```

```python
centroids
```

```python
avg_distance
```

```python
# Get the groups (clusters) and distances
from scipy.cluster.vq import vq

groups, cdist = vq(salaries, centroids)

plt.scatter(salaries, np.arange(0,100), c=groups)
plt.xlabel('Salaries in EUR')
plt.ylabel('Indices')
plt.show()
```
